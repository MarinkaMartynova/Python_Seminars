```python
from sympy import *
from sympy.plotting import plot

x = Symbol('x')
y = -12*x**4*sin(cos(x))-18*x**3+5*x**2 + 10*x - 30

plot(y, (x,-100, 100)) 
print("Рис.1 - задача представлена графиком в диапазоне от -100 до 100")
plot(y, (x, 0, 1))
print("Рис.2 - частный случай в диапазоне от 0 до 1")
plot(y, (x, -1, 0))
print("Рис.3 - частный случай в диапазоне от -1 до 0")
plot(y, (x, -0.5, 0.5))
print("Рис.3 - частный случай в диапазоне от -1 до 0")
```
